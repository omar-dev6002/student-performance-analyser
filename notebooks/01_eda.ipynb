```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys, os

# Allow imports from src/
sys.path.append(os.path.join(os.getcwd(), '..'))
from src.data_loader import load_data, quick_inspect
from src.preprocessing import preprocess

sns.set_theme(style='whitegrid', palette='muted')
plt.rcParams['figure.dpi'] = 120
print('✅ Imports done')
```

## 1. Load & Inspect

```python
df = load_data()   # reads data/raw/student-mat.csv
quick_inspect(df)
```

## 2. Univariate Analysis

```python
# Distribution of final grade G3
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].hist(df['G3'], bins=20, edgecolor='white', color='steelblue')
axes[0].set_title('Distribution of Final Grade (G3)')
axes[0].set_xlabel('G3 Score')
axes[0].set_ylabel('Count')

axes[1].hist(df['studytime'], bins=4, edgecolor='white', color='coral')
axes[1].set_title('Study Time Distribution')
axes[1].set_xlabel('Study Time (1=<2h, 4=>10h)')

plt.tight_layout()
plt.savefig('../outputs/figures/univariate.png', bbox_inches='tight')
plt.show()
```

## 3. Bivariate Analysis — Grade vs Key Features

```python
fig, axes = plt.subplots(2, 2, figsize=(13, 9))

# Study time vs G3
sns.boxplot(x='studytime', y='G3', data=df, ax=axes[0, 0])
axes[0, 0].set_title('Study Time vs Final Grade')

# Failures vs G3
sns.boxplot(x='failures', y='G3', data=df, ax=axes[0, 1])
axes[0, 1].set_title('Past Failures vs Final Grade')

# Parental education (Medu) vs G3
sns.boxplot(x='Medu', y='G3', data=df, ax=axes[1, 0])
axes[1, 0].set_title("Mother's Education vs Final Grade")

# Internet access vs G3
sns.boxplot(x='internet', y='G3', data=df, ax=axes[1, 1])
axes[1, 1].set_title('Internet Access vs Final Grade')

plt.tight_layout()
plt.savefig('../outputs/figures/bivariate.png', bbox_inches='tight')
plt.show()
```

## 4. Correlation Heatmap

```python
# Use preprocessed (encoded) version for correlation
df_clean = preprocess(df, save=True)

plt.figure(figsize=(14, 10))
corr = df_clean.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))  # upper triangle mask
sns.heatmap(corr, mask=mask, annot=False, cmap='coolwarm',
            linewidths=0.5, vmin=-1, vmax=1)
plt.title('Feature Correlation Matrix')
plt.tight_layout()
plt.savefig('../outputs/figures/correlation_heatmap.png', bbox_inches='tight')
plt.show()

# Top correlations with G3
print('\nTop 10 features correlated with G3:')
print(corr['G3'].abs().sort_values(ascending=False).head(11))
```
